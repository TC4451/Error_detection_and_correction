```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

img_dir_path = './images/elp_1022_4/'


img1 = cv2.imread(img_dir_path + 'layer_1000.jpg')
img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)
img2 = cv2.imread(img_dir_path + 'layer_1002.jpg')
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY)


diff = cv2.subtract(img1, img2)
plt.imshow(diff)

print(np.mean(diff))
```
